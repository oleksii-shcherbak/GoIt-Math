Дано 4 точки: $M_0(0, 0, 0), M_1(1, \frac{1}{3}, 0), M_2(0, 2, \frac{1}{4}), M_3(\frac{1}{2}, \frac{1}{2}, 1)$.

Також дано 3 вектори, які задають паралелепіпед із зазначеними точками:
- $\overline{a}=\overline{M_0M_1}$
- $\overline{b}=\overline{M_0M_2}$
- $\overline{c}=\overline{M_0M_3}$

Порахуй координати векторів через задані точки.  
Порахуй об'єм, площу повної поверхні паралелепіпеда та кути між ребрами паралелепіпеда, координати решти вершин паралелепіпеда.



$
\overline{a} = M_1 - M_0 =
\left( \begin{array}{c}
1 - 0 \\
\frac{1}{3} - 0 \\
0 - 0
\end{array} \right) =
\left( \begin{array}{c}
1 \\
\frac{1}{3} \\
0
\end{array} \right)
\\[10pt]
\overline{b} = M_2 - M_0 =
\left( \begin{array}{c}
0 - 0 \\
2 - 0 \\
\frac{1}{4} - 0
\end{array} \right) =
\left( \begin{array}{c}
0 \\
2 \\
\frac{1}{4}
\end{array} \right)
\\[10pt]
\overline{c} = M_3 - M_0 =
\left( \begin{array}{c}
\frac{1}{2} - 0 \\
\frac{1}{2} - 0 \\
1 - 0
\end{array} \right) =
\left( \begin{array}{c}
\frac{1}{2} \\
\frac{1}{2} \\
1
\end{array} \right)
$

$
\overline{b} \times \overline{c} =
\left| \begin{array}{ccc}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
0 & 2 & \frac{1}{4} \\
\frac{1}{2} & \frac{1}{2} & 1
\end{array} \right|
= \left( \frac{7}{8}, \frac{1}{8}, -1 \right)
$

$
\overline{a} \cdot (\overline{b} \times \overline{c}) =
1 \times \frac{7}{8} + \frac{1}{3} \times \frac{1}{8} + 0 \times (-1)
= \frac{7}{8} + \frac{1}{24}
= \frac{22}{24}
= \frac{11}{12}
$

$
V = \left| \frac{11}{12} \right| = \frac{11}{12}
$

---

$
S_{ab} = |\overline{a} \times \overline{b}| =
\sqrt{ \left( \frac{1}{6} \right)^2 + \left( -\frac{1}{4} \right)^2 + (2)^2 }
= \sqrt{ \frac{1}{36} + \frac{1}{16} + 4 }
= \sqrt{ \frac{65}{36} }
\approx 1.342
$

$$
S_{ac} = |\overline{a} \times \overline{c}| =
\sqrt{ \left( \frac{1}{2} \right)^2 + \left( -1 \right)^2 + \left( \frac{1}{3} \right)^2 }
= \sqrt{ \frac{1}{4} + 1 + \frac{1}{9} }
= \sqrt{ \frac{25}{18} }
\approx 1.178
$$

$
S_{bc} = |\overline{b} \times \overline{c}| =
\sqrt{ \left( \frac{7}{8} \right)^2 + \left( \frac{1}{8} \right)^2 + (-1)^2 }
= \sqrt{ \frac{49}{64} + \frac{1}{64} + 1 }
= \sqrt{ \frac{114}{64} }
= \sqrt{ \frac{57}{32} }
\approx 1.333
$

$
S = 2 \times (S_{ab} + S_{ac} + S_{bc})
= 2 \times (1.342 + 1.178 + 1.333)
= 2 \times 3.853
= 7.706
$
